## Найти аппроксимирующий полином. 

Ниже описаны функции для построения аппроксимирующего полинома.

```python
def model1(X,k,Y):
    M = []
    for i in range(k+1):
        M.append(X**(k-i))
    M = np.vstack(M).T
    A=np.dot(np.linalg.inv(np.dot(M.T,M)),np.dot(M.T,Y))
    return A

import sympy as sp
x = sp.symbols('x') 
def mnk(A, arg):
    f=np.zeros(len(A))
    res=0
    for coef in list(A):
        res = res*x+coef
    F = []
    for i in (list(arg)):
        F.append(res.subs(x,i))     
    return F
```

Теперь находится оптимальная степень полинома для каждой функции.

```python
X1 = np.arange(13)

def opt(X, Y):
    k = 0
    min0 =  np.sum((Y - mnk(model1(X,0,Y), X)) **2)
    min1 = np.sum((Y - mnk(model1(X,1,Y), X)) **2)
    while min1 < min0:
        k += 1
        min0 =  np.sum((Y - mnk(model1(X,k,Y), X)) **2)
        min1 = np.sum((Y - mnk(model1(X,k+1,Y), X)) **2)
    return(k)

for i in range(1,30):
    Y = Array[i]
    n=opt(X1,Y)
    print('Оптимальная степень полинома для данных', file1.columns[i], n)
```

Построение графиков и аппроксимирующих полиномов для каждой функции.

```python
fig,ax = plt.subplots(29,1)
plt.rcParams['figure.figsize'] = (20, 180)
for i in range(1,30):
    Y = Array[i]
    n=opt(X1,Y)
    A = model1(X1,n,Y)
    f=mnk(A,X1)   
    ax[i-1].tick_params(axis = 'both', labelsize = 14)
    ax[i-1].set_title('Аппроксимация', fontsize = 13)
    ax[i-1].set_xlabel('Дата', fontsize = 11)
    ax[i-1].set_ylabel('Значения', fontsize = 11)
    ax[i-1].plot(Array[0], Y, 'r', color = 'blue',label = file1.columns[i], markersize=1) 
    ax[i-1].plot(Array[0], f, 'r', color = 'green', markersize=1)
    ax[i-1].legend()
```

## 5. Определить точность аппроксимирующего полинома (значение СКО).

Среднеквадратическое отклонение рассчитывается по формуле $ \sqrt {\frac{1}{n}\sum_{i=1}^n(X_i-X)^2} $.

```python
def sco(k,X,Y):
    e=[]
    summ = 0
    for i in range(1,k+1):
        summ = (Y - mnk(model1(X,i,Y), X)) **2
        e.append(((1/len(Y))*np.sum(summ))**(1/2))
    return e
```

Рассчитаем СКО для каждого полинома.

```python
fig,ax = plt.subplots(29,1)
plt.rcParams['figure.figsize'] = (20, 180)
for i in range(1,30):
    Y = Array[i]
    n=opt(X1,Y)
    s=sco(n,X1,Y)
    num=np.arange(1,len(s)+1)  
    ax[i-1].tick_params(axis = 'both', labelsize = 14)
    ax[i-1].set_title(' ', fontsize = 13)
    ax[i-1].set_xlabel('Степень полинома', fontsize = 11)
    ax[i-1].set_ylabel('Ошибка', fontsize = 11)
    ax[i-1].plot( num, s, 'r', color = 'blue', label = file1.columns[i], markersize=1) 
    ax[i-1].legend()
```
